```python
#all_df= pd.get_dummies(all_d,drop_first=True)
from sklearn.preprocessing import LabelEncoder

def encoding_categorical_features(df):
    le = LabelEncoder()
    df['MonthOfMatch'] = le.fit_transform(df['MonthOfMatch'])
    df['Team1_Innings'] = le.fit_transform(df['Team1_Innings'])
    df['Team1_Venue'] = le.fit_transform(df['Team1_Venue'])
    df['Team2_Innings'] = le.fit_transform(df['Team2_Innings'])
    df['Team2_Venue'] = le.fit_transform(df['Team2_Venue'])
encoding_categorical_features(all_d)
```

```python
x=all_d.iloc[:2508,:]
y=df['MatchWinner']

test=all_d.iloc[2508:,:]
```

```python
def dummy_teams(df):
    dum=df[['Team1','Team2']]
    dum=dum.astype(str)
    dumy=dum.stack().str.get_dummies().sum(level=0)
    dumy=dumy.astype(int)
    df=pd.concat([df,dumy],1)
    df.drop(['Team1','Team2'],1,inplace=True)
    return df
```

```python
#x=pd.get_dummies(x['HostCountry'],prefix='HostCountry',)
#test=pd.get_dummies(test['HostCountry'],prefix='HostCountry',)
```

```python
x_1=dummy_teams(x)
#x_1=dummy_Host_country(x_1)
```

```python
test_1=dummy_teams(test)
#test_1=dummy_Host_country(test_1)
```

```python
dfh=pd.get_dummies(x_1['HostCountry'],prefix='HostCountry',)
x_1=pd.concat([x_1,dfh],1)
x_1.drop(['HostCountry'],1,inplace=True)
```

```python
dfh=pd.get_dummies(test_1['HostCountry'],prefix='HostCountry',)
test_1=pd.concat([test_1,dfh],1)
test_1.drop(['HostCountry'],1,inplace=True)
```

```python
for i in range(submission.shape[0]):
    team1=test.iloc[i]['Team1']
    team2=test.iloc[i]['Team2']
    if submission.iloc[i][team1] > submission.iloc[i][team2]:
        submission.iloc[i][team1]=1-submission.iloc[i][team2]
    else:
        submission.iloc[i][team2]=1-submission.iloc[i][team1]
```

```python
#nullfy the probability values
submission[submission<0.05]=0
submission[submission>0.95]=1
```
